This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")

df = data.copy()
```

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
df.dropna(how="all", inplace=True)
```

```python
df.drop("unnamed:_0", axis=1, inplace=True)
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
df.columns
```

```python
def morethan1000(total_claim_amount):
    return df[df[total_claim_amount]>1000]
```

```python
plus1k = morethan1000("total_claim_amount")
```

```python
plus1k
```

```python
plus1k_final = plus1k.groupby("response").get_group("Yes")
plus1k_final.sample(5)
```

```python
lessthan1k = plus1k_final[plus1k_final["customer_lifetime_value"] < 1000]
lessthan1k.shape
```

```python
morethan1k = plus1k_final[plus1k_final["customer_lifetime_value"] > 1000]
morethan1k.shape
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
df_yes = df.groupby("response").get_group("Yes")
df_yes.sample(3)
```

```python
policy_gender = df_yes.groupby(["policy_type","gender"])
```

```python
average_policy_gender = policy_gender["total_claim_amount"].mean()
average_policy_gender.reset_index()
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
df["number_of_policies"].value_counts()
```

```python
policies_state_group = df.groupby(["number_of_policies","state"])
```

```python
policies_state = policies_state_group["number_of_policies"].agg(["count"])
```

```python
policies_state.reset_index()
```

```python
morethan500 = policies_state[policies_state["count"] > 500]
```

```python
morethan500.reset_index()
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
education_gender_group = df.groupby(["education","gender"])
```

```python
education_gender= education_gender_group["customer_lifetime_value"].agg(["max","min","median"]).rename(columns={"max":"clv max","min":"clv min","median":"clv median"}).round(2)
```

```python
education_gender.reset_index()
```
